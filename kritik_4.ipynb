```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent_1d(func, derivative_func, learning_rate, start_point, num_iterations):
    points = [start_point]
    
    for _ in range(num_iterations):
        gradient = derivative_func(start_point)
        start_point = start_point - learning_rate * gradient
        points.append(start_point)

    return np.array(points)

def funny_function(x):
    if x > 0:
        return x**x
    elif x == 0:
        return 1
    else:
        return abs(x)**abs(x)

def derivative_funny_function(x):
    if x > 0:
        return x**x * (np.log(x) + 1)
    elif x == 0:
        return 0
    else:
        return abs(x)**abs(x) * np.log(abs(x)) * np.sign(x)

# Convert functions to handle scalar values
scalar_funny_function = np.vectorize(funny_function)
scalar_derivative_funny_function = np.vectorize(derivative_funny_function)

def plot_gradient_descent(func, points):
    x_vals = np.linspace(min(points) - 2, max(points) + 2, 100)
    y_vals = func(x_vals)

    plt.plot(x_vals, y_vals, label="Function to Minimize")
    plt.scatter(points, [func(p) for p in points], color='red', label="Gradient Descent")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.title("Gradient Descent for funny_function")
    plt.show()

# Example usage
learning_rate = 0.1
start_point = 1.5
num_iterations = 50

gradient_points = gradient_descent_1d(scalar_funny_function, scalar_derivative_funny_function, learning_rate, start_point, num_iterations)
plot_gradient_descent(scalar_funny_function, gradient_points)
```
