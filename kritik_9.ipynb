```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# a - define the function f(x, y) = e^x(sin(y))+y^3. Compute the first partial derivatives.

x, y = sp.symbols ('x y') # define symbols

f = sp.exp(x)*sp.sin(y)+y**3 # define function

df_dx = sp.diff (f, x) # differentiate with respect to x

df_dy = sp.diff (f, y) # differentiate with respect to y

print ("(a.i) The first partial derivative with respect to x is", df_dx, ".") # print results
print ("(a.ii) The first partial derivative with respect to y is", df_dy, ".")

# b - given the function g(x, y) = x^2y + xy^2, find the gradient vector and compute its magnitude at the point (1, âˆ’1).

g = x**2*y + x*y**2

dg_dx = sp.diff (g, x) # differentiate with respect to x

dg_dy = sp.diff (g, y) # differentiate with respect to y

grad_v = (dg_dx, dg_dy) # determine gradient vector

print ("                               ") 
print ("(b.i) The gradient vector equals", grad_v, ".") # print gradient vector results

p_d_x = dg_dx.subs ([(x,1),(y,-1)]) # calulate gradient vector x-coordinate @ (1,-1)

p_d_y = dg_dy.subs ([(x,1),(y,-1)]) # calulate gradient vector y-coordinate @ (1,-1)

print ("(b.ii) The gradient vector at (1,-1) is", p_d_x, ",", p_d_y, ".") # print gradient vector @ (1,-1)

mag = ((p_d_x)**2 + (p_d_y)**2)**(1/2) # calculate magnitude of gradient vector

print ("(b.iii) The magnitude of the gradient vector at (1, -1) is", round(mag, 3), ".") # print magnitude

# c - For the function h(x, y) = ln(x^2 +y^2), find the second partial derivatives. Discuss the symmetry of mixed partial derivatives.

h = sp.log(x**2 + y**2) # define function

dh_dx = sp.diff (h, x) # differentiate with respect to x - first partial derivative

dh_dy = sp.diff (h, y) # differentiate with respect to y - first partial derivative

d2h_dx2 = sp.simplify (sp.diff (dh_dx, x)) # differentiate with respect to x - second partial derivative

d2h_dy2 = sp.simplify (sp.diff (dh_dy, y)) # differentiate with respect to y - second partial derivative

d2h_dxdy = sp.simplify (sp.diff (dh_dy, x)) # differentiate with respect to x - mixed partial derivative

d2h_dydx = sp.simplify (sp.diff (dh_dx, y)) # differentiate with respect to y - mixed partial derivative

print ("                               ") 
print ("(c.i) The second partial derivative with respect to x is", d2h_dx2, ".") # print results
print ("(c.ii) The second partial derivative with respect to y is", d2h_dy2, ".")
print ("(c.iii) The mixed parial derivative with respect to x is", d2h_dxdy, ".")

if d2h_dxdy == d2h_dydx: # chheck if mixed partial derivatives are symmetrical or not
    print ("(c.iv) The mixed partial derivatives are symmetrical.") #print results
else:
    print ("(c.iv) The mixed partial derivatives are not symmetrical.")

# d  - create a contour plot of the function j(x, y) = x^3âˆ’3xy+y^3.

j = x**3 - 3*x*y + y**3 # define function using sympy

j_func = sp.lambdify((x,y),j,'numpy') # convert sympy expression to numerical function

x_vals = np.linspace(-3, 3, 400) # create grid of x and y values
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y) # evaluate j on grid

print ("                               ") 
print ("(d) The contour plot of $j(x, y) = x^3âˆ’3xy+y^3$.") # label results

plt.contourf(X, Y, Z, levels=50, cmap='viridis') # plot contour
plt.colorbar()
plt.title('Contour plot of j(x, y) = x^3âˆ’3xy+y^3')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# e -  create a contour plot a function of your choice!

i = (x**2 + y - 11)**2 + (x + y**2 - 7)**2 # define function using sympy

i_func = sp.lambdify((x,y),i,'numpy') # convert sympy expression to numerical function

x_vals = np.linspace(-4, 4, 400) # create grid of x and y values
y_vals = np.linspace(-4, 4, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = i_func(X, Y) # evaluate i on grid

print ("                               ") 
print ("(e) The contour plot of Himmelblau's function - i(x,y) = (i^2+y-11)^2+(x+y^2-7)^2.") # label results

plt.contourf(X, Y, Z, levels=80, cmap='twilight_shifted') # plot contour
plt.colorbar()
plt.title('Contour plot of Himmelblaus Function - $i(x,y) = (i^2+y-11)^2+(x+y^2-7)^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
